```python
n = [5,3,2,2,1,5,57,5,10]
m = [10,111,1,9,5,67,2]
freq_n = {}
for i in n:
    if i in freq_n:
        freq_n[i] += 1
    else:
        freq_n[i] = 1

for j in m:
    if j in freq_n:
        print(j, "is present", freq_n[j], "times")
    else:
        print(j, "is not present")
        




 

```

```python
## brute force approach 

n = [5,3,2,2,1,5,57,5,10]
m = [10,111,1,9,5,67,2]
for j in m:
    count = 0
    for i in n:
        if i == j:
            count += 1
    if count > 0:
        print(j, "is present", count, "times")
    else:
        print(j, "is not present")
```

```python
# hashlist approach
n = [5,3,2,2,1,5,57,5,10]
m = [10,111,1,9,5,67,2]
hashlist = [0]*11 # constraint of n is 1 to 10
for i in n:
    if 1 <= i <= 10:
        hashlist[i] += 1

for j in m:
    if j < 1 or j > 10:
        print(j, "is not present")
    else:
        print(j, "is present", hashlist[j], "times")
```
