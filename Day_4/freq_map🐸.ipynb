# Frequency Map

### used 2 pointers approach 

first if -> checks if the number is already in freq dict and is repeated then its freq is incremented , if new then gets introduced 

```python
nums = [5,6,7,7,1,9,111,1,1,5,1,1]
freq = {}
L = 0
R = len(nums) - 1
while L <= R:
    if nums[L] in freq:
        freq[nums[L]] += 1
    else:
        freq[nums[L]] = 1
    L += 1

print(freq)
```

##### way-2

same as before just using loop instead of pointers 

```python
nums = [5,6,7,7,1,9,111,1,1,5,1,1]
freq = {}
for i in range(0, len(nums)):
    if nums[i] in freq:
        freq[nums[i]] += 1
    else:
        freq[nums[i]] = 1
print(freq)
```

#### way-3

```python
nums = [5,6,7,7,1,9,111,1,1,5,1,1]
n = len(nums)
hash_map = {}
for i in range(n):
  hash_map[nums[i]] = hash_map.get(nums[i], 0) + 1
print(hash_map)

#hash_map.get(5, 0)  # returns 0 if 5 is not present in the hash_map else returns its value
```
