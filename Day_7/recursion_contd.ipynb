```python
def func(counter, n):
    if counter < n:
        print("Hello, World!")
        func(counter + 1, n)

func(0, 10)  # Will print 10 times
```

### PRINT 1 to N using recursion

```python
def func(N):
    if N > 0:
        func(N - 1)
        print(N)
func(5)  # Will print numbers from 1 to 5


    

```

```python
def func(i,n):
    if i>n:
        return
    print(i)
    func(i+1,n)
func(1,5)
```

```python
# backtracking print 1 to n
def func(n):
    if n==0:
        return
    func(n-1)
    print(n)
func(5)
```

```python
def func_forward(n):
    if n == 0:
        return
    print(n)        
    func_forward(n-1)

func_forward(5)
```

```python
def func(i,n):
    if i>n:
        return
    func(i+1,n)
    print(i)
func(1,5)
```
