## Right rotate an array by 1 place

```python
nums = [5,-2,3,9,0,6,10,7]
n = len(nums)

temp = nums[n-1]

for i in range(n-2,-1,-1):
    nums[i+1] = nums[i]

nums[0] = temp


print(nums)

```

## Rotate by k-places

```python
nums = [5,-2,3,9,0,6,10,7]
k = 3

n = len(nums)
rotations = k%n

for _ in range(0,rotations):
    e = nums.pop()
    nums.insert(0,e)

nums
```

```python
nums = [5,-2,3,9,0,6,10,7]
k = 3

n = len(nums)

k = k % n

nums[:] = nums[n - k :] + nums[: n- k]

nums
```

```python
nums = [5,-2,3,9,0,6,10,7]
k = 3
n = len(nums)

def reverse(nums,left,right):
    while left<right:
        nums[left],nums[right] = nums[right],nums[left]
        left+=1
        right -=1


reverse(nums,n-k,n-1)
reverse(nums,0,n-k-1)
reverse(nums,0,n-1)


nums

```
