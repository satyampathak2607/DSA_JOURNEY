# Print smth 4 times using recursion

```python

def greet():
    counter = 0
    print("Hello, World!")
    if counter <= 5: 
        
      
        greet()
        counter += 1
        return
    ########## EPIC FAIL ############

```

## we used counter as a parameter instead of a global variable 

```python
def greet(counter=0):
    print("Hello, World!")
    if counter < 5:
        greet(counter + 1)
greet()
```

```python
count = 0
def func():
    global count
    if count == 4:
        return
    print("saty")
    count += 1
    func()
func()
 # we used global variable ...for it u need to declare it as global inside the function
```
