#### FIND THE FACTORIAL OF THE NUMBER

```python
def fact(N):
    if N ==0:
        return 1
    return N * fact(N-1)

fact(1298)
```

### Reverse an array

```python
num = [5,7,3,2,6,1,5,9]
def r_a(num):
    L=0
    R=len(num)-1
    while L<R:
        num[L],num[R] = num[R],num[L]
        L+=1
        R-=1
    return num
   
r_a(num)

```

### using recursion

```python
num = [1,2,3,4,5]
def r_a(num,L,R):
    if L>=R:
        return
    num[L],num[R] = num[R],num[L]
    r_a(num,L+1,R-1)
    return num
r_a(num,0,len(num)-1)
```
