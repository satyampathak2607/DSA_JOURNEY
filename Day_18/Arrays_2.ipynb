## Check if an Array is Sorted

```python
nums = [1, 4,6,6,8]

L =0
n = len(nums)
for i in range(L+1, n):
    if nums[L]> nums[i]:
        print("unsorted")
        break
    L+=1
else:
    print("sorted")
```

```python
nums = [1, 4,6,6,8,7]
def check(nums):
    n = len(nums)
    for i in range(0,n-1):
        if nums[i]> nums[i+1]:
            return False
    return True

check(nums)
```

```python
nums = [1, 4,6,6,8,7]
if nums == sorted(nums):
    print("sorted")
else:
    print("unsorted")
```

##  Remove Duplicates from a Sorted Array

```python
nums =[0,0,1,1,1,1,2,3,6,6,9,9]
L=[nums[0]]
for r in range(1,len(nums)):
    if nums[r]!= nums[r-1]:
        L.append(nums[r])

print(L)
print(len(L))

```

```python
nums =[0,0,1,1,1,1,2,3,6,6,9,9]
def d(nums):
    i = 0
    for j in range(1,len(nums)):
        if nums[i] != nums[j]:
            i+=1
            nums[i] = nums[j]
    return nums[:i+1]

d(nums)


```
